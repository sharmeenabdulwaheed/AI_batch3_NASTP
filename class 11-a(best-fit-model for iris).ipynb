```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv("iris.csv")
```

## data balanced or not

```python
sns.countplot(data=data, x="species")
```

```python
from sklearn.neighbors import KNeighborsClassifier #it is a model
```

## we will create a variable to store our model, creating an object for our variable

```python
model=KNeighborsClassifier(n_neighbors=3)
#by default, number of neighbors are 5
#change neighbors and check for better accuracy
```

```python
features=data[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
```

```python
labels=data["species"]
```

```python
model.fit(features, labels)
#model creates a graph and give different color coding to different variables
```

```python
from sklearn.model_selection import train_test_split #it is functions
```

```python
X_train, X_test, y_train, y_test=train_test_split(features, labels, test_size=0.25)
#25% data will go in test data
#75% data will go in train data
```

## training of model will use only X_train and y_train

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
predictions=model.predict(X_test)
#model predictions
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(predictions, y_test)
```

## finding the best fit model

```python
neighbors=range(1,21)
accuracies=[]

for i in neighbors:
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    acc=knn.score(X_test, y_test)
    accuracies.append(acc)
```

```python
accuracies
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbors, accuracies)
```

```python
model=KNeighborsClassifier(n_neighbors=8)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

## storing model for later use

```python

import pickle
with open ("best_model.pkl", "wb") as file:
    pickle.dump(model, file)
    
```
