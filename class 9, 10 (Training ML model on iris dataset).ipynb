```python
from sklearn.neighbors import KNeighborsClassifier #it is a model
```

## we will create a variable to store our model, creating an object for our variable

```python
model=KNeighborsClassifier()
```

```python
features=data[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
```

```python
features
```

```python
labels=data["species"]
```

```python
model.fit(features, labels)
#model creates a graph and give different color coding to different variables
```

```python
model.predict([[2.3, 3.0,4.5, 1.2]])
```

```python
model.predict([[2.3, 3.0,4.5, 1.2]])[0]
```

```python
from sklearn.model_selection import train_test_split #it is functions
```

```python
X_train, X_test, y_train, y_test=train_test_split(features, labels, test_size=0.25)
#25% data will go in test data
#75% data will go in train data
```

## training of model will use only X_train and y_train

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
predictions=model.predict(X_test)
#model predictions
```

```python
y_test
#actual and correct answers
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(predictions, y_test)
```
